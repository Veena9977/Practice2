```python
#1.Histogram showing distribution values in the diabetes pedigree function column.
plt.hist(df['DiabetesPedigreeFunction'],rwidth=0.90)
plt.title('Diabetes')
plt.xlabel('DiabetesPedigreeFunction')
plt.ylabel('Frequency')
plt.show()
```

```python
#2.Scatter plot of Glucose level versus Insulin
plt.scatter(df["Glucose"],df["Insulin"])
plt.xlabel("Glucose")
plt.ylabel("Insulin")
plt.show()
```

```python
#Scatter plots showing relationship between Age and DiabetesPedigreeFunction and second scatter plot showing relationship between Age and Glucose levels.
plt.figure(figsize=(8,6))
plt.subplot(2,1,1)
plt.scatter(df["Age"],df["DiabetesPedigreeFunction"],c="black")
plt.xlabel("Age")
plt.ylabel("DiabetesPedigreeFunction")
plt.subplot(2,1,2)
plt.scatter(df["Age"],df["Glucose"],c="blue")
plt.xlabel("Age")
plt.ylabel("Glucose")
plt.show()
```

```python
#Heat map of columns Glucose,Age,Insulin and BloodPressure
import seaborn as sns
data = df[['Glucose','Age','Insulin','BloodPressure']]
plt.figure(figsize=(8,6))
sns.heatmap(data.corr(),annot=True)
plt.show()
```

```python
#Bar plot showing comparison of Age and BMI
plt.bar(df["Age"],df["BMI"])
plt.xlabel("Age")
plt.ylabel("BMI")
plt.show()
```

```python
#Pie plot showing diabetes outcome distribution
name = ['Without Diabetes','With Diabetes']
count= df["Outcome"].value_counts()
plt.pie(count,labels= name ,autopct="%1.1f")
plt.show()
```
